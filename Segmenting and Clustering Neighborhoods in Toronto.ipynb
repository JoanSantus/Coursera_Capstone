# Import Libraries

```python
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup
```

# Get webpage URL

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = urllib.request.urlopen(url)
soup = BeautifulSoup(page, "lxml")
print(soup.prettify())
```

# Search for table inside HTML

```python
all_tables=soup.find_all("table") 
all_tables
```

```python
right_table=soup.find('table', class_='wikitable sortable')
right_table
```

# Get Data into Tables

```python
A=[]
B=[]
C=[]
for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

# Create Dataframe and fill it with the Data inside the Tables


```python
df=pd.DataFrame(A,columns=['PostalCode'])
df['Borough']=B
df['Neighbourhood']=C
df
```

# Replace the values Not assigned
#### After some struggles I noticed it is "Not assigned\n"

```python
df= df[df['Borough'] != 'Not assigned\n'].reset_index(drop=True)
df
```

# Group by Postal Code and then sort them


```python
df.groupby("PostalCode")
df
```

```python
df= df.sort_values('PostalCode').reset_index(drop=True)
df
```

# Import Libraries

```python
import numpy as np
#!conda install -c conda-forge geopy --yes
from geopy.geocoders import Nominatim
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
#!conda install -c conda-forge folium=0.5.0 --yes
import folium
```

# Get Toronto's Latitude and Longitude

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="Toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

# I duplicate my dataframe to create 2 maps later

```python
df_merge2 = df_merge
```

# I create a map with all the Boroughs

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, df_merge2 in zip(df_merge2['Latitude'], df_merge2['Longitude'], df_merge2['Borough'], df_merge2['Neighbourhood']):
    label = '{}, {}'.format(df_merge2, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
